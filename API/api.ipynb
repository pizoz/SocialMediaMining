Import e creazione dell'header necessario per effettuare richiesta all'API

```python
import httpx
import json

with open("token.txt", "r") as f:
    token = f.read().strip()
    
with open("user_id.txt", "r") as f:
    user_id = f.read().strip()

with open("guild_id.txt", "r") as f:
    guild_id = f.read().strip()

with open("channel_id.txt", "r") as f:
    channel_id = f.read().strip()

with open("message_id.txt", "r") as f:
    message_id = f.read().strip()

with open("poll_id.txt", "r") as f:
    poll_id = f.read().strip()

headers = {
    "Authorization": token,
    "Content-Type": "application/json"
}
```

Richiesta dell'oggetto User corrispondente all'utente loggato

```python
endpoint = "https://discord.com/api/v9/users/@me"
request = httpx.get(endpoint, headers=headers)
response = request.json()
print(json.dumps(response, indent=4))
```

Richiesta di un oggetto user

```python
endpoint = f"https://discord.com/api/v9/users/{user_id}"
request = httpx.get(endpoint, headers=headers)
response = request.json()
print(json.dumps(response, indent=4))
```

Richiesta dei server a cui partecipa l'utente loggato

```python
endpoint = "https://discord.com/api/v9/users/@me/guilds"
request = httpx.get(endpoint, headers=headers)
response = request.json()
response
```

Richiesta dell'oggetto guild

```python
endpoint = f"https://discord.com/api/v9/guilds/{guild_id}"
request = httpx.get(endpoint, headers=headers)
response = request.json()
print(json.dumps(response, indent=4))
```

Richiesta dei membri di una guild

```python
endpoint = f"https://discord.com/api/v9/guilds/{guild_id}/members?limit=1000"
request = httpx.get(endpoint, headers=headers)
response = request.json()
with open('members_RealTropea.json', 'w') as f:
    json.dump(response, f, indent=4)
```

Richiesta dei canali presenti all'interno di una guild

```python
endpoint = f"https://discord.com/api/v9/guilds/{guild_id}/channels"
request = httpx.get(endpoint, headers=headers)
response = request.json()
response[2]
```

Richiesta dei messaggi di un canale (o thread)

```python
endpoint = f"https://discord.com/api/v9/channels/{channel_id}/messages"
request = httpx.get(endpoint, headers=headers)
response = request.json()
with open('response.json', 'w') as f:
    json.dump(response[0], f, indent=4)
print(json.dumps(response[0], indent=4))
```

Richiesta degli utenti che hanno risposto ad un poll

```python
endpoint = f"https://discord.com/api/v9/channels/{channel_id}/polls/{poll_id}/answers/1"
request = httpx.get(endpoint, headers=headers)
response = request.json()
print(json.dumps(response, indent=4))
endpoint = f"https://discord.com/api/v9/channels/{channel_id}/polls/{poll_id}/answers/2"
request = httpx.get(endpoint, headers=headers)
response = request.json()
print(json.dumps(response, indent=4))
```

Richiesta di un oggetto messaggio

```python
endpoint = f"https://discord.com/api/v9/channels/{channel_id}/messages/{message_id}"
request = httpx.get(endpoint, headers=headers)
response = request.json()
print(json.dumps(response, indent=4))
with open('response2.json', 'w') as f:
    json.dump(response, f, indent=4)
```
